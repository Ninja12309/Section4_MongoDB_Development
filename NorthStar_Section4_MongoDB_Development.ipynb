# Create collections

```python
customers_col = db["customers"]

orders_col = db["orders"]

drivers_col = db["drivers"]

vehicles_col = db["vehicles"]

deliveries_col = db["deliveries"]

complaints_col = db["complaints"]

app_events_col = db["app_events"]

incidents_col = db["incidents"]

hubs_col = db["hubs"]

print("Collections created")
```

# Insert JSON documents (CREATE)

```python
customers_col.insert_one({
"customer_id":"C0001",
"age":34,
"home_zone":"North",
"customer_type":"Consumer",
"loyalty_score":72,
"app_engagement_score":65,
"account_status":"Active"
})
```

# Insert multiple documents

```python
orders_col.insert_many([
{
"order_id":"O1001",
"customer_id":"C0001",
"service_type":"Parcel",
"pickup_zone":"North",
"dropoff_zone":"Central",
"priority_level":"High",
"order_value":45
},

{
"order_id":"O1002",
"customer_id":"C0002",
"service_type":"Passenger",
"pickup_zone":"South",
"dropoff_zone":"Airport",
"priority_level":"Medium",
"order_value":120
}
])
```

# READ operations

```python
for doc in orders_col.find():
    print(doc)
```

# Query with filter

```python
high_priority = orders_col.find(
{"priority_level":"High"}
)

for doc in high_priority:
    print(doc)
```

# UPDATE operation

```python
orders_col.update_one(

{"order_id":"O1001"},

{"$set":
{"order_value":55}

})
```

# DELETE operation

```python
orders_col.delete_one(

{"order_id":"O1002"}

)
```

# Aggregation example

```python
pipeline = [

{
"$group":
{
"_id":"$service_type",
"avg_value":
{"$avg":"$order_value"}
}

}

]

result = orders_col.aggregate(pipeline)

for r in result:
    print(r)
```

Average complaint resolution time by severity

Business purpose

Helps identify whether serious complaints take longer to resolve, indicating inefficiencies in customer service processes.

```python
# Aggregation 2
# Average complaint resolution time by severity

pipeline_resolution = [

{
"$group":
{
"_id":"$severity",

"avg_resolution_days":
{"$avg":"$resolution_days"},

"total_complaints":
{"$sum":1}
}

},

{
"$sort":
{"avg_resolution_days":-1}
}

]

result_resolution = complaints_col.aggregate(
pipeline_resolution
)

for r in result_resolution:
    print(r)
```
